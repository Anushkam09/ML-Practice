```python
from numpy import *
import operator
from os import listdir
```

```python
def classify0(X, dataSet, labels, k):
    dataSetSize = dataSet.shape[0]
    #calculate euclidean distance
    diffMat = tile(X, (dataSetSize,1)) - dataSet
    sqDiffMat = diffMat**2
    sqDistances = sqDiffMat.sum(axis = 1)
    distances = sqDistances**0.5

    #sorting and getting indices
    sortedDistIndices = distances.argsort()

    classCount = {}
    for i in range(k):
        votelabel = labels[sortedDistIndices[i]]
        classCount[votelabel] = classCount.get(votelabel, 0) + 1
    sortedClassCount = sorted(classCount.items(), key = operator.itemgetter(1), reverse=True)  #sort by values
    return sortedClassCount[0][0]
```

```python
def imagetovector(filename):
    return_vector = zeros((1,1024))
    fr = open(filename)
    for i in range(32):
        line = fr.readline()
        for j in range(32):
            return_vector[0, 32*i + j] = int(line[j])
    return return_vector
            
```

```python
def handwritingClassTest():
    labels = []
    file_list_training = listdir("digits/trainingDigits")
    m = len(file_list_training)
    trainingMat = zeros((m, 1024))
    for i in range(m):
        filename = file_list_training[i]
        fileStr= filename.split('.')[0]
        num = int(fileStr.split('_')[0])
        labels.append(num)
        trainingMat[i, :] = imagetovector(f"digits/trainingDigits/{filename}")

    error = 0.0
    file_list_test = listdir("digits/testDigits")
    mTest = len(file_list_test)
    for i in range(mTest):
        filename = file_list_test[i]
        fileStr= filename.split('.')[0]
        num = int(fileStr.split('_')[0])
        test_vector = imagetovector(f"digits/testDigits/{filename}")
        result = classify0(test_vector,trainingMat, labels, 3)
        print(f"The classifier came back with: {result}, the real answer is: {num}")
        if result != num:
            error += 1.0

    print(f"\nTotal error rate: {error / float(mTest):.2%}")    
        
```

```python
def classifyDigit():
    to_predict_file_name = input("Enter relative file path to predict digit: ")
    labels = []
    file_list_training = listdir("digits/trainingDigits")
    m = len(file_list_training)
    trainingMat = zeros((m, 1024))
    for i in range(m):
        filename = file_list_training[i]
        fileStr= filename.split('.')[0]
        num = int(fileStr.split('_')[0])
        labels.append(num)
        trainingMat[i, :] = imagetovector(f"digits/trainingDigits/{filename}")

    predict_vector = imagetovector(to_predict_file_name)
    result = classify0(predict_vector,trainingMat, labels, 3) 

    print("Predicted class:", result) 
```

```python
handwritingClassTest()
```
